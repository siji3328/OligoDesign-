```python
# 목표 DNA 서열
target_sequence = "ATGAGCAAAGGTGAAGAACTGTTTACCGGCGTTGTGCCGATTCTGGTGGAACTGGATGGCGATGTGAACGGTCACAAATTCAGCGTGCGTGGTGAAGGTGAAGGCGATGCCACGATTGGCAAACTGACGCTGAAATTTATCTGCACCACCGGCAAACTGCCGGTGCCGTGGCCGACGCTGGTGACCACCCTGACCTATGGCGTTCAGTGTTTTAGTCGCTATCCGGATCACATGAAACGTCACGATTTCTTTAAATCTGCAATGCCGGAAGGCTATGTGCAGGAACGTACGATTAGCTTTAAAGATGATGGCAAATATAAAACGCGCGCCGTTGTGAAATTTGAAGGCGATACCCTGGTGAACCGCATTGAACTGAAAGGCACGGATTTTAAAGAAGATGGCAATATCCTGGGCCATAAACTGGAATACAACTTTAATAGCCATAATGTTTATATTACGGCGGATAAACAGAAAAATGGCATCAAAGCGAATTTTACCGTTCGCCATAACGTTGAAGATGGCAGTGTGCAGCTGGCAGATCATTATCAGCAGAATACCCCGATTGGTGATGGTCCGGTGCTGCTGCCGGATAATCATTATCTGAGCACGCAGACCGTTCTGTCTAAAGATCCGAACGAAAAACGGGACCACATGGTTCTGCACGAATATGTGAATGCGGCAGGTATTACG"

# 올리고 설계 파라미터
oligo_length = 60  # 각 올리고뉴클레오타이드의 길이
overlap_min_tm = 52  # 중첩 영역의 최소 melting temperature (°C)
overlap_max_tm = 54  # 중첩 영역의 최대 melting temperature (°C)

def calculate_tm(sequence):
    """간단한 Wallace Rule을 사용하여 Tm 계산"""
    at_count = sequence.count('A') + sequence.count('T')
    gc_count = sequence.count('G') + sequence.count('C')
    return 2 * at_count + 4 * gc_count

def design_oligos(target_sequence, oligo_length, overlap_min_tm, overlap_max_tm):
    oligos = []
    start = 0
    overlap_length = 0

    while start < len(target_sequence):
        # 올리고 서열 추출
        end = start + oligo_length
        oligo = target_sequence[start:end]

        # 적절한 Tm을 가진 중첩 영역 찾기
        for overlap_length in range(20, oligo_length):  # 최소 20bp 중첩부터 시작
            overlap_sequence = target_sequence[end-overlap_length:end]
            tm = calculate_tm(overlap_sequence)
            if overlap_min_tm <= tm <= overlap_max_tm:
                break

        # 올리고 길이가 목표 길이보다 짧을 경우 패딩
        if len(oligo) < oligo_length:
            oligo = oligo.ljust(oligo_length, 'N')

        oligos.append(oligo)
        start += (oligo_length - overlap_length)  # 중첩된 영역만큼 이동

    return oligos

# 올리고 설계
oligos = design_oligos(target_sequence, oligo_length, overlap_min_tm, overlap_max_tm)

# 결과 출력
print(f"총 {len(oligos)}개의 올리고뉴클레오타이드가 생성되었습니다:")
for i, oligo in enumerate(oligos):
    print(f"Oligo {i+1}: {oligo}")
```
